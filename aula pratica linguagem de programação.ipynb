# NUMPY - 7º Ordenação e Adicionar Elementos

```python
import numpy as np

rng = np.random.default_rng()

array_1= rng.random([2,3])
print(f'Array 1 Sem ordenação:\n{array_1}\n')
print(f'Array 1 Sem ordenação:\n{np.sort(array_1, axis = 0)}\n')

# Concatenação

array_2 = rng.random([2,1])
print(f'Array 2:\n{array_2}\n')

Array_concatenado = np.concatenate([array_1,array_2], axis=1)
print(f'Array Concatenado:\n{Array_concatenado}\n')

array_concatenado_2 = np.hstack([array_1,array_2])
print(f'Array Concatenado:\n{array_concatenado_2}\n')
```

# NUMPY 9 - Como mudar o formato do array

```python
import numpy as np

array_a = np.arange(1,20)
print(f'Array_a:\n{array_a}\n')
array_a = array_a.reshape(4,5)
print(f'Array_a:\n{array_a}\n')

array_a = np.reshape(array_a, [4,5])
print(f'Array_a:\n{array_a}\n')

array_a = array_a.flatten()
print(f'Array_a:\n{array_a}\n')

array_b = array_a.reshape(4,5)
print(f'Array_b:\n{array_b}\n')

array_b = np.vsplit(array_b,4)
print(f'Array_b:\n{array_b}\n')
print(array_b[0])
```

# PLOTLY - 2º Criando arrays pra montar

```python
import numpy as np

a = 3/4
b = 2/4
c = 1/4
vetor_a = np.ones(20)
vetor_b = np.ones(20)
vetor_c = np.ones(20)

for i in range (1,20):
    vetor_a[i] = vetor_a[i -1] + a**i
    vetor_b[i] = vetor_b[i -1] + b**i
    vetor_c[i] = vetor_c[i -1] + c**i

print(vetor_a)
print(vetor_b)
print(vetor_c)

np.savetxt('vetor_a.txt', vetor_a, fmt='%f', delimiter= ';')
np.savetxt('vetor_b.txt', vetor_b, fmt='%f', delimiter= ';')
np.savetxt('vetor_c.txt', vetor_c, fmt='%f', delimiter= ';')
```

# PLOTLY - 3º Criando nosso primeiro gráfico

```python
import numpy as np
import plotly.express as px

array_a = np.loadtxt('vetor_a.txt', dtype = np.float64, delimiter=';')
array_b = np.loadtxt('vetor_b.txt', dtype = np.float64, delimiter=';')
array_c = np.loadtxt('vetor_c.txt', dtype = np.float64, delimiter=';')
print(array_a)


array_abc = np.vstack([array_a,array_b,array_c])
print(array_abc)
array_abc = array_abc.transpose()
print(array_abc)
fig = px.line(array_abc)
fig.show()
```
